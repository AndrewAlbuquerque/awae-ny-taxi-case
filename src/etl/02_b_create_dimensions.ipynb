Using the oficial data dictionaries (https://www.nyc.gov/assets/tlc/downloads/pdf/data_dictionary_trip_records_yellow.pdf and https://www.nyc.gov/assets/tlc/downloads/pdf/data_dictionary_trip_records_green.pdf), I choose to create 3 new lookup dimensions for vendorID, ratecodeID and paymentType, to be used at enrichment of the data at golden layer to be consumed by end users at streamlit.

```python
s3_ext_location = 's3://awae-yellow-taxi-case/'

from pyspark.sql.types import StructType, StructField, LongType, StringType
```

```python
# 1. table: tpep_provider
provider_schema = StructType([
    StructField("VendorID", LongType(), False),
    StructField("VendorDesc", StringType(), False)
])
provider_data = [
    (1, "Creative Mobile Technologies, LLC"),
    (2, "Curb Mobility, LLC"),
    (6, "Myle Technologies Inc"),
    (7, "Helix")
]
df_provider = spark.createDataFrame(provider_data, schema=provider_schema)
df_provider.write.format("delta").mode("overwrite").save(f"{s3_ext_location}/silver/dim_tpep_provider")
```

```python
# 2. table: ratecode
ratecode_schema = StructType([
    StructField("RatecodeID", LongType(), False),
    StructField("RatecodeDesc", StringType(), False)
])
ratecode_data = [
    (1, "Standard rate"),
    (2, "JFK"),
    (3, "Newark"),
    (4, "Nassau or Westchester"),
    (5, "Negotiated fare"),
    (6, "Group ride"),
    (99, "Null/unknown")
]
df_ratecode = spark.createDataFrame(ratecode_data, schema=ratecode_schema)
df_ratecode.write.format("parquet").mode("overwrite").save(f"{s3_ext_location}/silver/dim_ratecode")
```

```python
# 3. tablename: payment_type
payment_schema = StructType([
    StructField("payment_type", LongType(), False),
    StructField("payment_desc", StringType(), False)
])
payment_data = [
    (0, "Flex Fare trip"),
    (1, "Credit card"),
    (2, "Cash"),
    (3, "No charge"),
    (4, "Dispute"),
    (5, "Unknown"),
    (6, "Voided trip")
]
df_payment = spark.createDataFrame(payment_data, schema=payment_schema)
df_payment.write.format("parquet").mode("overwrite").save(f"{s3_ext_location}/silver/dim_payment_type")
```
